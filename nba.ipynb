```python
per_game_df = pd.read_csv('Player Per Game.csv')
```

```python
per_game_df.columns
```

```python
merged_df = pd.merge(df, per_game_df, on='player', how='inner')
merged_df.head()
```

```python
df['made_all_nba'] = (df['type'] == 'All-NBA').astype(int)
```

```python
merged_df['made_all_nba'] = (merged_df['type'] == 'All-NBA').astype(int)
```

```python
merged_df.columns
```

```python
train_data = merged_df[merged_df['season_x'].between(2015, 2020)]

test_data = merged_df[merged_df['season_x'] >= 2021]
```

```python
features = [
    'fg_percent', 'x3p_percent','pts_per_game', 'ast_per_game',
    'trb_per_game','stl_per_game', 'blk_per_game', 'fga_per_game',
    'tov_per_game','fta_per_game', 'e_fg_percent', 'x3pa_per_game'
]

X_train = train_data[features]
y_train = train_data['made_all_nba']

X_test = test_data[features]
y_test = test_data['made_all_nba']
```

```python
# = X.fillna(0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

model = RandomForestClassifier(random_state=42)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
importances = model.feature_importances_

for feature, importance in zip(features, importances):
    print(f"{feature}: {importance:.4f}")
```

```python
merged_df = pd.merge(merged_df, advanced_df, left_on=['player', 'season_y'], right_on=['player', 'season'], how='inner')
```

```python
merged_df.columns
```

```python
merged_df = merged_df.drop_duplicates(subset=['player', 'season_x'])
```

```python
merged_df = merged_df.dropna(axis=1, how='all')
```

```python
print(merged_df.index.is_unique)
```

```python
train_data = merged_df[merged_df['season_x'].between(2015, 2020)]

test_data = merged_df[merged_df['season_x'] >= 2021]


X_train = train_data[features]
y_train = train_data['made_all_nba']

X_test = test_data[features]
y_test = test_data['made_all_nba']

print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)
```

```python
features = [
    'x3p_percent','pts_per_game', 'ast_per_game',
    'trb_per_game', 'vorp',
    'tov_per_game','fta_per_game','usg_percent',
     'f_tr', 'ts_percent', 'ast_percent', 'ws',
]

X_train = train_data[features]
y_train = train_data['made_all_nba']

X_test = test_data[features]
y_test = test_data['made_all_nba']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

model = RandomForestClassifier(random_state=42)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
importances = model.feature_importances_

for feature, importance in zip(features, importances):
    print(f"{feature}: {importance:.4f}")
```
